```python
import plotly.express as px
import pandas as pd


```

```python

air = pd.read_csv('./data/air_quality.csv')
climate = pd.read_csv('./data/climate.csv')
co2 = pd.read_csv('./data/co2.csv')
pollution = pd.read_csv('./data/pollution.csv')
```

```python
def show_map(df, value, title="Sample Map"):
    fig = px.choropleth(df,
                        locations='country_code',
                        color=value,
                        hover_name='country_code',
                        color_continuous_scale=px.colors.sequential.Plasma,
                        title=title)

    fig.show()
```

## Air Quality Data

```python
show_map(air, 'avg')
```

## Climate Data

```python

climate_2008_aug = climate[(climate['year'] == 2008) & (climate['month'] == 8)]
show_map(climate_2008_aug, 'temp_mean')
```

## Co2 Data

```python
co2_2019 = co2[co2['year'] == 2019]
show_map(co2_2019, 'co2_per_capita')
```

## Pollution Data

```python
pollutant = "PM2.5"

poll = pollution[pollution['pollutant']==pollutant]

show_map(poll, "value")
```
